```python
import pandas as pd
import cleaning as cl

# Load and clean dataset
marketing_customer = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')
clean_marketing_customer = cl.clean_data(marketing_customer)

# -------------------------------
# Task 1: Pivot Table - Total Revenue by Sales Channel
# -------------------------------
pivot_total_revenue = pd.pivot_table(
    clean_marketing_customer,
    values='total_claim_amount',        # assuming total_claim_amount = total revenue
    index='sales_channel',
    aggfunc='sum'
)
print("Total Revenue by Sales Channel:")
print(pivot_total_revenue)

# -------------------------------
# Task 2: Pivot Table - Avg. Customer Lifetime Value per Gender and Education
# -------------------------------
pivot_avg_clv = pd.pivot_table(
    clean_marketing_customer,
    values='customer_lifetime_value',
    index='gender',
    columns='education',
    aggfunc='mean'
)
print("\nAverage Customer Lifetime Value per Gender and Education:")
print(pivot_avg_clv)
```
